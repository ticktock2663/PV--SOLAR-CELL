```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('ts_20211126.csv')
df
```

```python
sensor01 = []
sensor02 = []
manaDat = []

for i in range(0,len(df),30):
    sensor01 += [df['sensor01'][i]]
    sensor02 += [df['sensor02'][i]]
    manaDat += [df['manaDat'][i]]
```

```python
for i in range(len(manaDat)):
    manaDat[i] = str(manaDat[i])[:12]
print(manaDat)
```

```python
df2=pd.DataFrame([manaDat,sensor01, sensor02])
df2
```

```python
df2=df2.transpose()
df2
```

```python
df2=df2.rename(columns={0:'date',1:'temp',2:'humi'})
df2
```

```python
df2.index = pd.to_datetime(df2['date'])
df2=df2.drop('date',axis=1)
df2
```

```python
df2.to_csv('210610.csv')
```
