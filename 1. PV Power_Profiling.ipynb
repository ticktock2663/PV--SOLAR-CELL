#### 5) 계절별 발전량 프로파일

```python
#봄철 인덴스
df2.index.month == 3 |(df2.index.month == 4)|(df2.index.month ==5 )
```

```python
#봄철 인덴스
spring_index =(df2.index.month == 3) |(df2.index.month == 4)|(df2.index.month == 5 )
#여름철 인덴스
summer_index =(df2.index.month == 6) |(df2.index.month == 7)|(df2.index.month == 8 )
#가을철 인덴스
fall_index =(df2.index.month == 9) |(df2.index.month == 10)
#겨울철 인덴스
winter_index =(df2.index.month == 11) |(df2.index.month == 12)|(df2.index.month == 1 )|(df2.index.month == 2) 
```

```python
df2[spring_index].mean().plot(figsize=(15,4),fontsize=13, xticks = df2.columns, style = '.-')
df2[summer_index].mean().plot(figsize=(15,4),fontsize=13, xticks = df2.columns, style = '.-')
df2[fall_index].mean().plot(figsize=(15,4),fontsize=13, xticks = df2.columns, style = '.-')
df2[winter_index].mean().plot(figsize=(15,4),fontsize=13, xticks = df2.columns, style = '.-')
plt.grid()
plt.legend(['봄','여름','가을','겨울'])
plt.xlabel('Hour',fontsize=13)
plt.ylabel('Power [Kw]',fontsize=13)
plt.title('2021년 계절별 발전량 패턴',fontsize=13)
```
